# MLB Jersey RAG System - Demo Notebook

This notebook demonstrates how to use the MLB Jersey RAG system for semantic search.

## 1. Setup and Initialization

```python
import sys
sys.path.append('..')

from src.search import JerseySearch

# Initialize the search engine
searcher = JerseySearch(
    data_path='../data/jerseys.json',
    persist_directory='../chroma_db'
)

print("Search engine initialized!")
```

## 2. Initialize Database (First Time Only)

```python
# Only run this once to initialize the database
searcher.initialize_database(reset=False)
```

## 3. Basic Search Examples

```python
# Search for pinstripe jerseys
results = searcher.search("pinstripe jerseys", top_k=3)
searcher.print_results(results)
```

```python
# Search for red alternate jerseys
results = searcher.search("red alternate jerseys", top_k=5)
searcher.print_results(results)
```

```python
# Search by description
results = searcher.search("classic traditional design with blue colors", top_k=5)
searcher.print_results(results)
```

## 4. Advanced Queries

```python
# Search for throwback/vintage style jerseys
results = searcher.search("throwback vintage retro cream colored jerseys", top_k=5)
searcher.print_results(results)
```

```python
# Search for unique color combinations
results = searcher.search("unique unusual color scheme jersey", top_k=5)
searcher.print_results(results)
```

## 5. Filter by Team or Color

```python
# List all teams
teams = searcher.list_teams()
print(f"Total teams: {len(teams)}\n")
for team in teams[:10]:
    print(f"  • {team}")
```

```python
# Get all jerseys for a specific team
yankees_jerseys = searcher.filter_by_team("New York Yankees")
print(f"New York Yankees Jerseys: {len(yankees_jerseys)}\n")
for jersey in yankees_jerseys:
    print(f"  {jersey['type']}: {jersey['description'][:100]}...")
```

```python
# Filter by color
blue_jerseys = searcher.filter_by_color("blue")
print(f"Found {len(blue_jerseys)} jerseys with blue\n")
for jersey in blue_jerseys[:5]:
    print(f"  {jersey['team']} - {jersey['type']}")
```

## 6. Explore Search Results

```python
# Get detailed results for analysis
results = searcher.search("orange jersey", top_k=5)

import pandas as pd

# Convert to DataFrame for easy viewing
df = pd.DataFrame(results)
df[['team', 'type', 'primary_color', 'secondary_color', 'score']]
```
